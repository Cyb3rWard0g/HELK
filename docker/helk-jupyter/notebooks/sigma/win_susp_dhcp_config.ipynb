# DHCP Server Loaded the CallOut DLL
This rule detects a DHCP server in which a specified Callout DLL (in registry) was loaded

## Rule Content
```
- title: DHCP Server Loaded the CallOut DLL
  id: 13fc89a9-971e-4ca6-b9dc-aa53a445bf40
  status: experimental
  description: This rule detects a DHCP server in which a specified Callout DLL (in
    registry) was loaded
  references:
  - https://blog.3or.de/mimilib-dhcp-server-callout-dll-injection.html
  - https://technet.microsoft.com/en-us/library/cc726884(v=ws.10).aspx
  - https://msdn.microsoft.com/de-de/library/windows/desktop/aa363389(v=vs.85).aspx
  date: 2017/05/15
  author: Dimitrios Slamaris
  tags:
  - attack.defense_evasion
  - attack.t1073
  logsource:
    product: windows
    service: system
    category: null
  detection:
    selection:
      EventID: 1033
    condition: selection
  falsepositives:
  - Unknown
  level: critical

```

## Querying Elasticsearch

### Import Libraries

````
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Search
import pandas as pd
````

### Initialize Elasticsearch client

````
es = Elasticsearch(['http://helk-elasticsearch:9200'])
searchContext = Search(using=es, index='logs-endpoint-winevent-system-*', doc_type='doc')
````

### Run Elasticsearch Query

````
s = searchContext.query('query_string', query='event_id:"1033"')
response = s.execute()
if response.success():
    df = pd.DataFrame((d.to_dict() for d in s.scan()))
````
